<h3 style="color: pink">Tiền xử lý dữ liệu</h3>

---

Tiến hành loại bỏ các thuộc tính không liên quan tới câu hỏi, chỉ giữ lại các thuộc tính để phân tích như `Producers`, `Studios`, `Score`, `Popularity`, `Favorites`, `Type`

```python
filtered_df = anime_clean_df[['Producers', 'Studios', 'Score', 'Popularity', 'Favorites', 'Type']]
filtered_df
```

Kiểm tra các cột có giá trị `UNKNOWN` hoặc `Unknown`

```python
# Kiểm tra các cột có giá trị Unknown hoặc UNKNOWN
unknown_columns = []

for column in filtered_df.columns:
    if filtered_df[column].isin(['Unknown', 'UNKNOWN']).any():
        unknown_columns.append(column)

# In ra danh sách các cột có giá trị Unknown hoặc UNKNOWN
print("Các cột có giá trị 'Unknown' hoặc 'UNKNOWN':")
print(unknown_columns)
```

- Loại bỏ hoàn toàn các dòng mà cả cột `Studios` và `Producers` đều có giá trị là 'UNKNOWN'. Những dòng này không cung cấp bất kỳ thông tin hữu ích nào cho việc phân tích, giữ lại chúng chỉ làm tăng độ nhiễu của dữ liệu.
- Thay thế 'UNKNOWN' bằng chuỗi rỗng nếu chỉ có một trong hai cột có giá trị 'UNKNOWN', việc giữ lại các dòng này giúp tận dụng tối đa thông tin có sẵn, đặc biệt là khi một trong hai cột chứa thông tin hữu ích.

```python
# Xóa các dòng nếu cả hai cột đều có giá trị 'UNKNOWN'
filtered_df = filtered_df[~((filtered_df['Studios'] == 'UNKNOWN') & (filtered_df['Producers'] == 'UNKNOWN'))]

# Thay thế 'UNKNOWN' bằng chuỗi rỗng nếu chỉ có một trong hai cột có giá trị 'UNKNOWN'
filtered_df['A'] = filtered_df['Studios'].replace('UNKNOWN', '')
filtered_df['B'] = filtered_df['Producers'].replace('UNKNOWN', '')
```

Kết hợp hai cột `Producers` và cột `Studios` để tạo ra cột `Producer-Studio` nhằm thể hiện quan hệ giữa nhà sản xuất và studio

```python
filtered_df['Producer-Studio'] = filtered_df['Producers'] + ' - ' + filtered_df['Studios']
```

- Nhóm dữ liệu theo từng cặp `Producer-Studio` để phân tích hiệu suất hợp tác của từng cặp.
- Tìm ra những cặp hợp tác thành công nhất dựa trên các chỉ số như `Score`, `Popularity` và `Favourity`

```python
# Group by 'Producer-Studio' and calculate mean values for Score, Popularity, and Favorites
partnership_performance = filtered_df.groupby('Producer-Studio')[['Score', 'Popularity', 'Favorites']].mean()

# Sort the results to find the best-performing partnerships
partnership_performance = partnership_performance.sort_values(by='Score', ascending=False)
```

<h3 style="color: pink">Khám phá dữ liệu - Trả lời câu hỏi</h3>

---

Tiến hành trực quan hóa hiệu suất của các cặp `Producer-Studio` hàng đầu dựa trên điểm trung bình của chúng

```python
# Plot the top 10 partnerships based on score
top_partnerships = partnership_performance.head(10)

# Create a bar plot for Score
top_partnerships['Score'].plot(kind='bar', figsize=(10, 6), color='skyblue')
plt.title('Top 10 Producer-Studio Partnerships by Score')
plt.xlabel('Producer-Studio Partnership')
plt.ylabel('Average Score')
plt.xticks(rotation=45, ha='right')
plt.show()
```

Tiến hành trực quan hóa số thể loại anime của các cặp `Producer-Studio` hàng đầu sản xuất

```python
# Filter the top-performing partnerships
top_partnerships_data = filtered_df[filtered_df['Producer-Studio'].isin(top_partnerships.index)]

# Analyze the Type distribution for top-performing partnerships
type_distribution = top_partnerships_data.groupby('Producer-Studio')['Type'].value_counts().unstack().fillna(0)

# Plot the type distribution
type_distribution.plot(kind='bar', stacked=True, figsize=(12, 8))
plt.title('Distribution of Anime Types by Producer-Studio Partnerships')
plt.xlabel('Producer-Studio Partnership')
plt.ylabel('Count of Anime Types')
plt.xticks(rotation=45, ha='right')
plt.show()
```

**Nhận xét**
- `Sunrise` cùng với `Bandai Namco Pictures` là hai studio tham gia vào sản xuất nhiều bộ anime thuộc top hàng đầu
- `Aniplex`, `TV Tokyo`, `Dentsu`, `Miracle Robo`, `Tokuma` là các nhà sản xuất có nhiều bộ anime thuộc top đầu
- Thể loại `TV` là thể loại có nhiều bộ anime thuộc top nhất
- Đa số sự kết hợp của `Aniplex, TV Tokyo - Sunrise` mang lại nhiều hiệu quả nhất
